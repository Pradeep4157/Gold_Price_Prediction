Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

Data Collection and Processing

```python
#Importing the csv Data to pandas Dataframe
gold_data=pd.read_csv('/content/gold price dataset.csv')
```

```python
#Printing the first 5 rows of the Dataframe
gold_data.head()
```

```python
#Printing the last 5 rows of the Dataframe
gold_data.tail()
```

```python
#number of rows and colums in the Dataframe
gold_data.shape
```

```python
# Getting some more information about the Dataframe
gold_data.info()
```

```python
#Checking the null values in the Dataframe
gold_data.isnull().sum()
```

```python
# Getting the statistical Data measures about the data
gold_data.describe()
```

Correlation:

1) Positive Correlation

2) Negative Correlation

```python
correlation = gold_data.select_dtypes(include=['number']).corr()
```

```python
plt.figure(figsize=(8,8));
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
sns.distplot(gold_data['GLD'],color='green')
```

Splitting Features and Target

```python
X=gold_data.drop(['GLD','Date'],axis=1)
Y=gold_data['GLD']
```

```python
print(X)
```

```python
print(Y)
```

Splitting into Training Data and Testing Data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
regressor=RandomForestRegressor(n_estimators=100)
```

```python
regressor.fit(X_train,Y_train)
```

Model Evaluation

```python
test_data_prediction=regressor.predict(X_test)
```

```python
error_score=metrics.r2_score(Y_test,test_data_prediction)
print("R-Squared Error : ",error_score)
```

Plotting the Actual Values and the Predicted Values

```python
Y_test=list(Y_test)
```

```python
plt.plot(Y_test,color='blue',label='Actual Value')
plt.plot(test_data_prediction,color='green',label='Predicted Value')
plt.title('Actual Price Vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Prices')
plt.legend()
plt.show()
```
